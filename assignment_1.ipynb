```python
#User login check
username = input()
password = input()

if username == 'admin' and password == '1234':
    print('Login successful')
else:
    print('Invalid credentials')
```

```python
#pass-fail analyzer
marks = [45, 78, 90, 33, 60]
p = f = 0
for i in marks:
    if i >= 50:
        p += 1
    else:
        f += 1
print("number of passed students:", p)
print("number of failed students:", f)
```

```python
#simple data cleaner
names = [" Alice ", "bob", " CHARLIE "]
for i in range(len(names)):
    names[i] = names[i].strip()
    names[i] = names[i].lower()
print(names)
```

```python
#message length analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    if len(i)>10:
        print('length of message is longer than 10 characters')
    else:
        print(len(i))
```

```python
#error message detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
e = 0
for i in logs:
    if i.lower() == 'error':
        e += 1
print('Error count:', e)
```
