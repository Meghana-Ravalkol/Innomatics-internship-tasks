```python
#Highest Salary from Employee Data
n = int(input("enter number of employees: "))
employees = {}
for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter employee salary: "))
    employees[name] = salary
hsal = 0
for key, value in employees.items():
    if hsal < value:
        hname = key
        hsal = value
print("Highest Salary:",hname,'-',hsal)
```

```python
# Find Maximum and Minimum Values
numbers = list(map(int, input().split()))
maxm = 0
minm = float('inf')
for i in numbers:
    if maxm < i:
        maxm = i
    if minm > i:
        minm = i
print("list:", numbers)
print("max:", maxm)
print("min:", minm)
```

```python
# count Products Above a Price Threshold
prices = list(map(int, input().split()))
c = 0
for i in prices:
    if i > 1000:
        c += 1
print("Products above 1000:", c)
```

```python
# Calculate Attendance Percentage
attendance = list(input().split())
c = 0
for i in attendance:
    if i == 'P':
        c += 1
print("Attendance percentage:", (c/len(attendance))*100)
```

```python
# Count Character Frequency
s = input()
d = {}
for i in s:
    if i != ' ':
        if i in d.keys():
            d[i] += 1
        else:
            d[i] = 1
print(d)
```

```python
# Check if a Key Exists in a Dictionary
sample_employees = {
    1: {"name": "Ravi", "salary": 75000}, 
    2: {"name": "Anita", "salary": 63000}, 
    3: {"name": "Kiran", "salary": 72000}
}
key = input()
for i in range(1, len(sample_employees)+1):
    if sample_employees[i]['name'] == key:
        print("Employee exists")
        break
else:
    print("Employee doesn't exist")
```
