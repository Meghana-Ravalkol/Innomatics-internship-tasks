```python
# Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
mp = 0
me = []
for e in employees:
    if employees[e] > mp:
        mp = employees[e]
for e in employees:
    if employees[e] == mp:
        me.append(e)
print(f"Top Performers Eligible for Bonus:",end=' ')
print(*me, sep=',',end=' ')
print(f"(Score: {mp})")
```

```python
# Search Query Keyword Analysis
s = "Buy mobile phone buy phone online".lower()
s = list(s.split())
d = {}
for i in s:
    if i in d.keys():
        d[i] += 1
    else:
        d[i] = 1
for i in list(d.keys()):
    if d[i] == 1:
        d.pop(i)
print(d)
```

```python
# Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
res = []
for i in range(len(sensor_readings)):
    if sensor_readings[i]%2 == 0:
        a = (i, sensor_readings[i])
        res.append(a)
print(res)
```

```python
 # Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
cg = cy = 0
for i in emails:
    a = i.split('@')
    if a[1] == 'gmail.com':
        cg += 1
    else:
        cy += 1
print(f'gmail.com: {(cg/len(emails))*100:.0f}%')
print(f'yahoo.com: {(cy/len(emails))*100:.0f}%')
```

```python
# Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales)/len(sales)
x = avg * 1.3
for i in range(len(sales)):
    if sales[i] > x:
        print(f"Day {i}: {sales[i]}")
```

```python
# Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
d = {}
for i in user_ids:
    if i in d.keys():
        d[i] += 1
    else:
        d[i] = 1
for i in d.keys():
    if d[i] > 1:
        print(f"{i}->{d[i]} times")
```
