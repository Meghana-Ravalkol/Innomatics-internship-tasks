```python
# Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    
    def turn_on(self):
        self.status = "ON"
    
    def turn_off(self):
        self.status = "OFF"
    
    def display_status(self):
        print(f"{self.name} is {self.status}")

light_name = input("enter light name: ")
action = input("enter action (ON/OFF): ")

light = SmartLight(light_name)
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
# Employee ID Card System
class EmployeeIDCardSys:
    def __init__(self, name, idnum, dept):
        self.name = name
        self.idnum = idnum
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.idnum)
        print("Department:", self.dept)

print("Enter details:")
ename = input("Employee Name: ")
emp_id = input("Emplyee ID: ")
dept = input("Department: ")

employee1 = EmployeeIDCardSys(ename, emp_id, dept)
employee1.display()
```

```python
# Mobile Contact Record
d = {}
class MobileContactRecord:
    def __init__(self, name, contact):
        self.name = name
        self.contact = contact
    def store_contact(self, name, contact): #storing contacts in a dictionary
        d[contact] = name
        print("Contact saved")
    def display(self):
        print("Name: ", self.name)
        print("Contact: ", self.contact)

name = input('Enter name: ')
contact = int(input('Enter contact: ')) 
contact1 = MobileContactRecord(name, contact) #can use loop to store multiple contacts
contact1.store_contact(name, contact)
contact1.display()
```

```python
# Product Price Tag Generator
class ProductPriceTagGenerator:
    def __init__(self, pname, price):
        self.pname = pname
        self.price = price

    def display(self):
        print("Product name: ", self.pname)
        print(f"Price: ₹{self.price}")

print("Enter details:")
pname = input("Product name: ")
price = input("Price: ")

product1 = ProductPriceTagGenerator(pname, price)
product1.display()
```

```python
# Movie Rating Display System
class MovieRatingDisplaySystem:
    def __init__(self, mname, rating):
        self.mname = mname
        self.rating = rating

    def display(self):
        print("Movie name: ", self.mname)
        print(f"Rating: {self.rating}/5")

mname = input("Enter movie name: ")
rating = float(input("Enter rating: "))

movie1 = MovieRatingDisplaySystem(mname, rating)
movie1.display()
```

```python
# Delivery Address Manager
class DeliveryAddressManager:
    def __init__(self, cname, addr):
        self.cname = cname
        self.addr = addr

    def display(self):
        print("Customer name: ", self.cname)
        print("Address:", self.addr)

cname = input("Enter customer name: ")
addr = input("Enter address: ")

print("Delivery details:")
cust1 = DeliveryAddressManager(cname, addr)
cust1.display()
```
